```python
import csv
import random
import time

# paramètres
num_rows = 10000  # change à 100k si besoin
start_timestamp = 1422629490813

output_file = "sensor_data.csv"

with open(output_file, mode='w', newline='') as file:
    writer = csv.writer(file)

    # header
    writer.writerow([
        "timestamp","AccelerationX","AccelerationY","AccelerationZ",
        "MagneticFieldX","MagneticFieldY","MagneticFieldZ",
        "Z-AxisAgle(Azimuth)","X-AxisAngle(Pitch)","Y-AxisAngle(Roll)",
        "GyroX","GyroY","GyroZ","session","location_id","device"
    ])

    timestamp = start_timestamp

    for i in range(num_rows):
        # simulation mouvement réaliste (bruit + variation douce)
        acc_x = random.uniform(-0.5, 0.5)
        acc_y = random.uniform(-9.0, -7.5)  # gravité dominante
        acc_z = random.uniform(5.5, 6.5)

        mag_x = random.uniform(2, 6)
        mag_y = random.uniform(30, 36)
        mag_z = random.uniform(0.3, 3)

        azimuth = random.uniform(340, 360)
        pitch = random.uniform(50, 55)
        roll = random.uniform(-2, 2)

        gyro_x = random.uniform(-0.1, 0.1)
        gyro_y = random.uniform(-0.3, 0.3)
        gyro_z = random.uniform(-0.3, 0.3)

        writer.writerow([
            timestamp, acc_x, acc_y, acc_z,
            mag_x, mag_y, mag_z,
            azimuth, pitch, roll,
            gyro_x, gyro_y, gyro_z,
            1, 34, "smartwatch"
        ])

        timestamp += 100  # 10 Hz

print(f"Dataset généré : {output_file}")
```
